# Description

This model uses results from Footywire to predict the results of AFL matches and the current season.

```python
from ipywidgets import Button
import pandas as pd
from afl import run,loop
def model(predict):
    pd.read_pickle('df')
    run()
predict = Button(description='Run model')
predict.on_click(model)
```
